```python
debut_count = {}
debut_sum = {}

def get_index_dan(rank):
    for i in range(len(rank)):
        if rank[i] in ['k', 'd']:
            return i
    return -1

def get_int_from_rank(rank):
    ind = get_index_dan(rank)
    if ind < 0 or rank[0] == 'P' or not rank[:ind].isdigit():
        return None
    if rank[ind] == 'k':
        return -int(rank[:ind]) + 1
    else:
        return int(rank[:ind])

for root, dirs, files in  tqdm(os.walk("../small_games/")):
    for file in tqdm(files):
        name = os.path.join(root, file)
        with open(name, "rb") as f:
            try:
                game = sgf.Sgf_game.from_bytes(f.read())
                root_node = game.get_root()
                w_rank = get_int_from_rank(root_node.get("WR"))
                b_rank = get_int_from_rank(root_node.get("BR"))
                moves_white = []
                moves_black = []
                for i, node in enumerate(game.get_main_sequence()):
                    if i == 0:
                        continue
                    if i >= 15:
                        break
                    
                    if i % 2 == 0:
                        moves_white.append(node.get_move()[1])
                    else:
                        moves_black.append(node.get_move()[1])
                moves_white = tuple(moves_white)
                moves_black = tuple(moves_black)
                
                debut_count.setdefault(moves_white, 0)
                debut_count.setdefault(moves_black, 0)
                debut_sum.setdefault(moves_white, 0)
                debut_sum.setdefault(moves_black, 0)
                
                debut_count[moves_white] += 1
                debut_count[moves_black] += 1
                debut_sum[moves_white] += w_rank
                debut_sum[moves_black] += b_rank   
                
            except ValueError:
                print(name)
```

```python
mx = 0
debut_rank = []
for key, val in debut_count.items():
    if val > 10:
        mean = float(debut_sum[key]) / float(val)
        debut_rank.append([key, mean])
```

```python
df = pd.DataFrame(debut_rank, columns=['moves', 'rank'])
df.sort_values(by='rank')
```
