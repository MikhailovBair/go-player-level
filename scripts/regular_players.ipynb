```python
from sgfmill import sgf
import os
import pandas as pd
from tqdm.notebook import tqdm
```

```python
table = []
for root, dirs, files in  tqdm(os.walk("../games/")):
    print(root)
```

```python
table = []
for root, dirs, files in  tqdm(os.walk("../games/")):
    for file in tqdm(files):
        name = os.path.join(root, file)
        with open(name, "rb") as f:
            game = sgf.Sgf_game.from_bytes(f.read())
            root_node = game.get_root()
            w_player = root_node.get("PW")
            row = [w_player, root, name]
            table.append(row)
```

```python
table
```

```python
df = pd.DataFrame(table, columns=['player', 'rank', 'filename'])
df.head()
```

```python
df.groupby(['player', 'rank']).count().sort_values(by='filename', ascending=False).head(1000)
```
